```python
import pandas as pd
import numpy as np
bj = pd.read_csv('project2/FiveCitiePMData/BeijingPM20100101_20151231.csv')
sh = pd.read_csv('project2/FiveCitiePMData/ShanghaiPM20100101_20151231.csv')
gz = pd.read_csv('project2/FiveCitiePMData/GuangzhouPM20100101_20151231.csv')
cd = pd.read_csv('project2/FiveCitiePMData/ChengduPM20100101_20151231.csv')
sy = pd.read_csv('project2/FiveCitiePMData/ShenyangPM20100101_20151231.csv')
```

```python
bj.drop(bj[gz['year'] < 2014].index, inplace = True)
sh.drop(sh[cd['year'] < 2014].index, inplace = True)
gz.drop(gz[gz['year'] < 2014].index, inplace = True)
cd.drop(cd[cd['year'] < 2014].index, inplace = True)
sy.drop(sy[sy['year'] < 2014].index, inplace = True)
```

```python
bj['average'] = bj.iloc[:,6:10].mean(axis = 1)
sh['average'] = sh.iloc[:,6:9].mean(axis = 1)
gz['average'] = gz.iloc[:,6:9].mean(axis = 1)
cd['average'] = cd.iloc[:,6:9].mean(axis = 1)
sy['average'] = sy.iloc[:,6:9].mean(axis = 1)
```

```python
bj = bj.drop(columns=['PM_Dongsi','PM_Dongsihuan','PM_Nongzhanguan', 'PM_US Post'])
bj.head()
```

```python
sh = sh.drop(columns=['PM_Jingan','PM_Xuhui', 'PM_US Post'])
sh.head()
```

```python
gz = gz.drop(columns=['PM_City Station','PM_5th Middle School', 'PM_US Post'])
gz.head()
```

```python
cd = cd.drop(columns=['PM_Caotangsi','PM_Shahepu', 'PM_US Post'])
cd.head()
```

```python
sy = sy.drop(columns=['PM_Taiyuanjie','PM_US Post','PM_Xiaoheyan'])
sy.head()
```

```python
bj['city'] = 'Beijing'
sh['city'] = 'Shanghai'
gz['city'] = 'Guangzhou'
cd['city'] = 'Chengdu'
sy['city'] = 'Shenyang'
```

```python
citypm = pd.concat([bj, sh, gz, cd, sy], axis=0)
citypm['season'] = citypm['season'].map({1:'Spring',2:'Summer',3:'Fall',4:'Winter'})
```
